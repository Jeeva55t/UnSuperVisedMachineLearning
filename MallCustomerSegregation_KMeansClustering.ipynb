```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df)
```

```python
df.drop('CustomerID',axis=1,inplace=True)
df.drop('Age',axis=1 , inplace = True)
df.drop('Genre',axis=1,inplace = True)
```

```python
df
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++')
y_kmeans = kmeans.fit_predict(df)
y_kmeans
```

```python
kmeans.inertia_
```

```python
df1 = df.values
plt.figure(figsize=(15,7))
plt.scatter(df1[y_kmeans==0,0],df1[y_kmeans==0,1],color='green',label='Clus1',s=50)
plt.scatter(df1[y_kmeans==1,0],df1[y_kmeans==1,1],color='blue',label='Clus2',s=50)
plt.scatter(df1[y_kmeans==2,0],df1[y_kmeans==2,1],color='red',label='Clus3',s=50)
plt.scatter(df1[y_kmeans==3,0],df1[y_kmeans==3,1],color='pink',label='Clus4',s=50)
plt.scatter(df1[y_kmeans==4,0],df1[y_kmeans==4,1],color='yellow',label='Clus5',s=50)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black',label='Centroids',s=150,marker=',')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
